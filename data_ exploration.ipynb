```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
days = df.여행일수.value_counts().sort_index()
days.index
```

```python
days = df.여행일수.value_counts().sort_index()

sns.barplot(x = days.index, y = days.values)#, palette='Set2')
plt.title('여행일수')
plt.xlabel('days')
plt.show()
```

```python
n_companion = df['동반 인원수'].value_counts().sort_index()

sns.barplot(x = n_companion.index, y = n_companion.values)
plt.xlabel('n_companion')
plt.show()
```

```python
companion_type = df['동반 여행 종류'].value_counts().sort_index()

sns.barplot(x = companion_type.index, y = companion_type.values)
plt.xlabel('companion_type')
plt.xticks(rotation=90)
plt.show()
```

```python
month = df['여행 시작 월'].value_counts().sort_index()

sns.barplot(x = month.index, y = month.values)
plt.title('월별 여행 횟수')
plt.xlabel('month')
plt.show()
```

```python
companion_type = df['동반 여행 종류'].value_counts().sort_index()

plt.pie(companion_type, labels=companion_type.index, autopct='%.1f%%')
plt.xlabel('companion_type')
plt.show()
```

```python
income = df['소득수준'].value_counts().sort_index()


ax = sns.barplot(x=income.index, y=income.values, palette='Set2')
total = float(len(df))
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height, f'{height / total:.1%}', ha='center', va='bottom')

plt.xlabel('income')
plt.xticks(rotation = 90)
plt.show()
```

```python
types = df['여행지 유형'].value_counts()


ax = sns.barplot(x=types.index, y=types.values, palette='Set2')
total = float(len(df))
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height, f'{height / total:.1%}', ha='center', va='bottom')

plt.xlabel('types')
plt.xticks(rotation = 90)
plt.show()
```

```python
# 성별 나이별 빈도 계산
freq_df = df.groupby(['연령대','성별']).size().unstack(fill_value=0)


# 그래프 생성
plt.figure(figsize=(10, 6))
sns.heatmap(freq_df, annot=True, cmap='Blues', fmt='.2f')
plt.title("Frequency by Gender and Ages")
plt.xlabel("Gender")
plt.ylabel("Ages")
plt.show()
```
